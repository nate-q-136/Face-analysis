```python
class Config:
    def __init__(self, input, output, detector_weights, checkpoint, with_persons, disable_faces, draw = False):
        self.input = input  # Update with your input path
        self.output = output  # Update with your output path
        self.detector_weights = detector_weights  # Update with detector weights path
        self.checkpoint = checkpoint  # Update with checkpoint path
        self.with_persons = with_persons
        self.disable_faces = disable_faces
        self.device = "cuda" if torch.cuda.is_available() else "cpu"
        self.draw = draw
yolo_path = "MiVOLO/yolov8x_person_face.pt"
model_path = "MiVOLO/model_imdb_cross_person_4.22_99.46.pth.tar"
# Initialize and set up configuration
config = Config(input, "output", yolo_path, model_path, 1, 0)
setup_default_logging()
def race_prediction(image_crop:np.ndarray, race_model):
    IMAGE_HEIGHT = 100
    IMAGE_WIDTH = 100
    ID_RACE_MAP = {0: 'caucasian', 1: 'mongoloid', 2: 'negroid'}

    resized_image = cv2.resize(image_crop, (IMAGE_HEIGHT, IMAGE_WIDTH))
    resized_image = np.array(resized_image).reshape((1, IMAGE_HEIGHT, IMAGE_WIDTH, 3))/255
    prediction = race_model.predict(resized_image)
    y_pred = ID_RACE_MAP[np.argmax(prediction)]

    return y_pred

def skintone_prediction(detected_face, model):
    classes = ['dark', 'mid-dark', 'mid-light', 'light']

    detected_face = cv2.resize(detected_face, (120, 90))
    detected_face = preprocess_input(detected_face[np.newaxis, ...])

    predictions = model.predict(detected_face)
    predicted_class_idx = np.argmax(predictions)
    predicted_class = classes[predicted_class_idx]
    return predicted_class

def mask_prediction(image_crop:np.ndarray, mask_net):
    face = cv2.cvtColor(image_crop, cv2.COLOR_BGR2RGB)
    face = cv2.resize(face, (224, 224))
    face = img_to_array(face)
    face = preprocess_input(face)

    face = np.expand_dims(face, axis=0)
    preds = mask_net.predict(face, batch_size=32)

    for pred in preds:
        mask, unmasked = pred
        label = "masked" if mask > unmasked else "unmasked"
        return label
def crop(image, bbox):
    """
    Crops a portion of an image specified by the bounding box.

    :param image: numpy array, image from which to crop
    :param bbox: tuple or list, bounding box with format (x1, y1, x2, y2)
    :return: numpy array, cropped image
    """
    x1, y1, x2, y2 = bbox
    # Ensure coordinates are within the image dimensions
    h, w = image.shape[:2]
    x1, x2 = max(0, x1), min(w, x2)
    y1, y2 = max(0, y1), min(h, y2)

    # Crop the image
    cropped_image = image[y1:y2, x1:x2]
    return cropped_image
```

```python
from google.colab.patches import cv2_imshow
from tqdm import tqdm
def main():
    if torch.cuda.is_available():
      torch.backends.cuda.matmul.allow_tf32 = True
      torch.backends.cudnn.benchmark = True
      predictor = Predictor(config, verbose=True)
      input_type = get_input_type(config.input)
      image_files = config.input
      # Model paths
      race_model_path = "models/race/race_classification_resnet50v2.h5"
      skintone_model_path = "models/skin_tone/skin_tone_100_epoch_model.h5"
      mask_model_path = "models/mask/mask_detector.model"
      emotion_model_path = "models/emotion/fer2013_mini_XCEPTION.110-0.65.hdf5"
      face_model_path = ""
      gender_model_path = ""
      age_model_path = ""

      # Load models
      race_model = load_model(race_model_path)
      skintone_model = load_model(skintone_model_path)
      mask_model = load_model(mask_model_path)
      for id, name in tqdm(enumerate(file_name)) :
      # Image path
        image_path = os.path.join(folder_path, name)
        image = cv2.imread(image_path)
        res_age, res_gender, detected_objects = predictor.recognize(image)
        face_bbox = detected_objects.get_bboxes_inds("face")
        for i in face_bbox :
          bbox = detected_objects.get_bbox_by_ind(i)
          # print(bbox)
          crop_img = crop(image, bbox)
          # print(crop_img.shape)
          # bbox[2] = bbox[2] - bbox[0]
          # bbox[3] = bbox[3] - bbox[1]
        # Race detection
          race = race_prediction(image_crop=crop_img, race_model=race_model)
          # Skintone detection
          skintone = skintone_prediction(detected_face=crop_img, model=skintone_model)
          # Mask detection
          mask = mask_prediction(image_crop=crop_img, mask_net=mask_model)
          emotion_handler = EmotionHandler()
          emotion_prediction = emotion_handler.emotions_predict(crop_img)
          # cv2_imshow(crop_img)
          data['file_name'].append(name)
          data['bbox'].append(bbox)
          data['image_id'].append(id+1)
          data['race'].append(race)
          data['age'].append(res_age[i - len(face_bbox)])
          data['emotion'].append(emotion_prediction)
          data['gender'].append(res_gender[i - len(face_bbox)])
          data['skintone'].append(skintone)
          data['masked'].append(mask)
          # print(res_age[i - len(face_bbox)], res_gender[i - len(face_bbox)])
          # print(emotion_prediction)
          # print(mask, skintone, race)


if __name__ == "__main__":
    main()
```
